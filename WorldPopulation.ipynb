# World Population Visualization

## Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Load Data

```python
df = pd.read_csv('/content/World Population Dataset.csv')
```

## Visualizatios of population per year

```python
df.columns
```

```python
non_year_columns = ['Rank', 'country', 'Capital', 'Continent', 'Area (km²)','World Population Percentage']
```

```python
format = {'1970 Population': '1970', '1980 Population': '1980','1990 Population': '1990',
                  '2000 Population': '2000','2010 Population': '2010','2015 Population': '2014',
                  '2020 Population': '2020'}
df = df.rename(columns=format)
```

```python
year_columns = df.columns.difference(non_year_columns)
year_columns
```

```python
df[year_columns].sum()
```

```python
plt.figure(figsize=(10, 6))
df[year_columns].sum().plot(kind='bar')
plt.title('World Population per year')
plt.xlabel('Year')
plt.ylabel('Popularion')
plt.xticks(rotation=0)
plt.grid(True)
plt.legend(title='Columns')
plt.show()
```

## Population per continent each year

```python
continent = df.groupby('Continent')[year_columns].sum()
continent
```

```python
for year in continent.columns:
  plt.figure(figsize=(8, 6))
  continent[year].plot(kind='bar')
  plt.title(f'World Population in {year}')
  plt.ylabel('Popularion')
  plt.xticks(rotation=0)
  plt.grid(True)
  plt.legend(title='Columns')
  plt.show()
```

## Population per country each year

```python
countries = df.groupby('country')[year_columns].sum()
```

```python
countries
```

## Area (km²) and population per country

```python
fig = px.scatter(df, x='Area (km²)', y='World Population Percentage', hover_data={'country': True})
fig.show()
```

## Population density by country in 2023

```python
density = df['2023']/df['Area (km²)']
density
```

```python
df['density in 2023'] = density
df
```

## Density of poblation by continent

```python
palette_color = sns.color_palette('bright')
data = df.groupby('Continent')['density'].sum()
plt.pie(data.values, labels=data.index, colors=palette_color, autopct='%.0f%%')
```
